```python
# q5
stu_info = {
    "Negi" : 9.08,
    "Bhandari": 9.23,
    "Mehul" : 8.89
}


while True:
    val = input("Enter the value : ")

    match val:
        case "A":
            name = input("Enter the name : ")
            cgpa = float(input("Enter Cgpa : "))
            new_item = {name : cgpa}
            stu_info.update(new_item)
            
        case "B":
            search_str = input("Enter the search name: ")
            updated_val = float(input("Enter new cgpa: "))
            stu_info[search_str] = updated_val

        
        case "C":
            search_str = input("Enter the search name: ")
            rslt = stu_info.get(search_str)
            if rslt is None:
                print("No such student exist")
            else:
                print(search_str,rslt)

        case "D":
            for key, val in stu_info.items():
                print(key,val)
        
        case _:
            break
```
