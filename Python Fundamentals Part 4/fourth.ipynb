```python
#q1 

class BankAccount: 
    def __init__(self,acc_number, owner_name, bal):
        self.acc_number = acc_number
        self.owner_name = owner_name
        self.bal = bal
    
    def deposit(self ,amt):
        self.bal += amt
        print(self.bal)

    def withdraw(self,amt):
        self.bal -= amt
        print(self.bal)

    def check_bal(self):
        print(f"The current balance is {self.bal}")



b1 = BankAccount(12345, "negi",1000000)
b1.deposit(100000)
b1.withdraw(5000)
b1.check_bal()
```

```python
#q2

class Book:
    def __init__(self, title, author, reviews):
        self.title = title
        self.author = author
        self.reviews = reviews

    def add_new_review(self, new_review):
        self.reviews.append(new_review)
        print(self.reviews)

    def count_reviews(self):
        print(f"The total reviews are {len(self.reviews)}")
    
    def display_all_reviews(self):
        for i in self.reviews:
            print(i)


b1 = Book("The Negi Tales", "Negi",["Best book","Great writer","Why not oscar yet"])
b1.add_new_review("Holy sht, best book & best writer")
b1.count_reviews()
b1.display_all_reviews()
```

```python
#q3
class Student:
    def __init__(self,name,roll_no,marks):
        self._name = name
        self._roll_num = roll_no
        self._marks = marks
    
    def getter(self):
        print(self._name," ",self._roll_num," ",self._marks)
    
    def setter(self,new_name, new_marks):
        if new_marks >= 0 and new_marks <=100: 
            self._marks = new_marks
        if new_name != "":
            self._name = new_name

s1 = Student("Negi", "21BDS0286", 9.08)
s1.setter("Best Negi",9.99)
s1.getter()
```

```python
# q4
class Shape:
    def area():
        pass

class Circle:
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        print(f"The area is {3.14 * (self.radius ** 2)}")

class Rectangle:
    def __init__(self, len, bdt):
        self.len = len
        self.bdt = bdt
    
    def area(self):
        print(f"The area is {self.len * self.bdt}")

c = Circle(10)
c.area()

r = Rectangle(10,10)
r.area()
```

```python
#q5
class Vehicle: 
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

class Car(Vehicle):
    def __init__(self,brand,model,seats):
        super().__init__(brand, model)
        self.seats = seats
    
    def getter(self):
        print(self.model, self.brand,self.seats)

class Bike(Vehicle):
    def __init__(self,brand,model,engine):
        super().__init__(brand, model)
        self.engine = engine
    
    def getter(self):
        print(self.model, self.brand,self.engine)

c = Car("MCZ", "Latest", "89")
c.getter()

b = Bike("MCZ", "Latest", 1000)
b.getter()
```

```python
# q6
from abc import ABC, abstractmethod

class Employee:
    @abstractmethod
    def calcuate_salary():
        pass

class Intern(Employee):
    def __init__(self, role):
        self.role = role
    def calcuate_salary(self):
        if self.role == "Intern":
            print(f"Your salary is {0.5 * 800000}")
        
class FullTimeEmployee(Employee):
    def __init__(self, role):
        self.role = role
    def calcuate_salary(self):
        if self.role == "Employee":
            print(f"Your salary is {1 * 800000}")
              
class ContractEmployee(Employee):
    def __init__(self, role):
        self.role = role
    def calcuate_salary(self):
        if self.role == "Contract":
            print(f"Your salary is {0.9 * 800000}")

i = Intern("Intern")
i.calcuate_salary()
f = FullTimeEmployee("Employee")
f.calcuate_salary()
c = ContractEmployee("Contract")
c.calcuate_salary()
```
