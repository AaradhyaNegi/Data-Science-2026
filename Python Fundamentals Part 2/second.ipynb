```python
## Practise qq 1
def factorial(x):
    if x == 0 | x == 1:
        return 1
    return x * factorial(x-1)

print(factorial(5))
```

```python
## Practise qq 2
def largest(a,b,c):# 1,2,3
    largest = 0
    if a > b and a > c:
        largest = a
    elif a < b and b > c:
        largest = b
    else:
        largest = c
    return largest


print(largest(1,2,3))
```

```python
# q1 
def tax_bracket(sal):
    bracket = ""
    if sal < 30_000:
        bracket = "5%"
    elif sal >= 30_000 and sal < 70_000:
        bracket = "15%"
    elif sal >= 70_000:
        bracket = "25%"
    else:
        return None
    return bracket

salary = 100_000
print(tax_bracket(salary))
```

```python
# q4 
def count_digits(n):
    cnt = 0
    while n >= 1:
        cnt += 1 
        n = int(n/10)
    return cnt

print(count_digits(12345))
```

```python
# q5 
def sum_of_digits(n):
    rslt = 0
    while n >=1:
        val = n % 10
        rslt += val
        n = int(n/10)

    return rslt

sum_of_digits(1234)
```

```python
# q8
def calculator(a, b, op):
    match op:
        case "+":
            print(a+b)
        case "-":
            print(a-b)
        case "*":
            print(a*b)
        case "/":
            print(a/b)
        case _:
            print("Enter valid numbers / operations")

calculator(1,2,"+")
calculator(3,4,"*")
calculator(a,b,":")
```

```python
# q9
def is_prime(n):
    isPrime = True
    if n == 1:
        print("Neither prime nor composite")
    
    for i in range(2, n):
        if n % i == 0:
            isPrime = False
            break
    return isPrime


if(is_prime(1245)):
    print("Prime")
else:
    print("Composite")
```

```python
# q10 
secret_num = 56

while True:
    guess = int(input("Enter your guess: "))
    rslt = (guess == secret_num)

    match rslt:
        case True:
            print("You won ! ")
            break
        case False: 
            if guess > secret_num:
                print("Too high !")
            else:
                print("Too Low !")
        case _:
            print("Enter valid data")
```
